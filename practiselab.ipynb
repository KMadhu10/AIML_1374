```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/individual+household+electric+power+consumption.zip")
```

```python
reasons = ['Heating and Cooling', 'Water Heating', 'Appliances', 'Lighting', 'Electronics']
consumption = [300, 200, 150, 100, 80]
plt.bar(reasons, consumption, color=['blue', 'orange', 'green', 'red', 'purple'])
plt.xlabel('Reasons for Electric Consumption')
plt.ylabel('Estimated Consumption (kWh)')
plt.title('Top 5 Reasons for Electric Consumption')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
